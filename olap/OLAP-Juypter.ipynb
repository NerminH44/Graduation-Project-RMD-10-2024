```python
query2 = '''
Select  
    O.order_id,
    O.order_date,
    O.SKU,
    O.quantity,
    O.shipment_id,
    O.retailer_id,
    O.employee_id,
    O.payment_id,
    O.order_status,
    O.quantity * P.unit_price as amount
from Orders O
left join Products P
on O.SKU = P.SKU
'''

fact_Orders = sqldf(query2, globals())


fact_Orders['date_key'] = pd.to_datetime(fact_Orders['order_date']).dt.strftime('%Y%m%d')


fact_Orders.head()
```

```python
import mysql.connector

# Establish a connection to MySQL
connection = mysql.connector.connect(
    host="localhost",         # or the server IP
    user="root",
    password="root",
    database="project"
)

# Create a cursor object
cursor = connection.cursor()

# SQL query to create the table
create_table_query = '''
CREATE TABLE Fact_Order (
    order_id INT NOT NULL,
    order_date DATE NOT NULL,
    SKU INT NOT NULL,
    quantity INT NOT NULL,
    shipment_id INT NOT NULL,
    retailer_id INT NOT NULL,
    employee_id INT NOT NULL,
    payment_id INT NOT NULL,
    order_status VARCHAR(255) NOT NULL,
    amount FLOAT DEFAULT NULL,
    date_key VARCHAR(8) NOT NULL,
    PRIMARY KEY (order_id)
);
'''

# Execute the query to create the table
cursor.execute(create_table_query)
print("Table 'Orders' created successfully.")

# Close the cursor and connection
cursor.close()
connection.close()
```

```python
import pandas as pd

# Define the empty DataFrames for each table
fact_orders = pd.DataFrame(columns=[
    'order_id', 'amount', 'date_key', 'order_date', 'SKU', 
    'quantity', 'shipment_id', 'retailer_id', 'employee_id', 
    'payment_id', 'order_status'
])

dim_products_manufacturers = pd.DataFrame(columns=[
    'SKU', 'product_name', 'description', 'stock_quantity', 
    'unit_price', 'manufacturer_id', 'manufacturer_name', 
    'phone_number', 'email', 'category_id'
])

dim_categories = pd.DataFrame(columns=[
    'category_id', 'category_name', 'Brand'
])

dim_department_jobs_managers = pd.DataFrame(columns=[
    'department_id', 'department_name', 'location_id', 
    'manager_id', 'manager_name', 'job_id', 'job_name'
])

dim_employees = pd.DataFrame(columns=[
    'employee_id', 'first_name', 'last_name', 'phone_number', 
    'email', 'salary', 'hire_date', 'department_id'
])

dim_retailers_locations = pd.DataFrame(columns=[
    'retailer_id', 'first_name', 'last_name', 'phone_number', 
    'email', 'location_id', 'location_name', 'city', 'postal_code'
])

dim_payments = pd.DataFrame(columns=[
    'payment_id', 'order_id', 'payment_method', 
    'payment_date', 'payment_amount'
])

dim_shipments = pd.DataFrame(columns=[
    'shipment_id', 'order_id', 'shipment_address', 
    'shipment_date', 'arrival_date', 'shipment_status', 
    'employee_id', 'retailer_id'
])

dim_inventory_warehouse = pd.DataFrame(columns=[
    'inventory_id', 'available_quantity', 'min_stock_level', 
    'max_stock_level', 'reorder_point', 'location_id', 
    'warehouse_id', 'warehouse_capacity', 'manager_id'
])

dim_date = pd.DataFrame(columns=[
    'date_id', 'date', 'date_key', 'day', 
    'month', 'month_name', 'quarter', 'year', 
    'day_of_week', 'day_of_week_name', 'week_of_year', 
    'is_weekend', 'fiscal_year', 'fiscal_quarter'
])

# Create a Pandas Excel writer using Openpyxl as the engine
with pd.ExcelWriter('data_export.xlsx', engine='openpyxl') as writer:
    fact_orders.to_excel(writer, sheet_name='Fact_Orders', index=False)
    dim_products_manufacturers.to_excel(writer, sheet_name='dim_products_manufacturers', index=False)
    dim_categories.to_excel(writer, sheet_name='dim_categories', index=False)
    dim_department_jobs_managers.to_excel(writer, sheet_name='dim_department_jobs_managers', index=False)
    dim_employees.to_excel(writer, sheet_name='dim_employees', index=False)
    dim_retailers_locations.to_excel(writer, sheet_name='dim_retailers_locations', index=False)
    dim_payments.to_excel(writer, sheet_name='dim_payments', index=False)
    dim_shipments.to_excel(writer, sheet_name='dim_shipments', index=False)
    dim_inventory_warehouse.to_excel(writer, sheet_name='dim_inventory_warehouse', index=False)
    dim_date.to_excel(writer, sheet_name='dim_date', index=False)

print("Data exported to 'data_export.xlsx' successfully!")
```
